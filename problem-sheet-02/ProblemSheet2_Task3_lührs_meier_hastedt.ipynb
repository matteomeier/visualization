```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

**subtask 1**

```python
imgs = loadmat("problem-sheet-02_proteins.mat")
```

```python
def show_img(img):
    fig, ax = plt.subplots()
    plt.imshow(img)
    ax.axis("off")
    plt.show()
```

```python
keys = ["A", "B", "C"]
for key_lt in keys:
    show_img(imgs[key_lt])
```

**subtask 2**

```python
#create image
black_img = np.zeros((imgs["A"].shape[0], imgs["A"].shape[1], 3)) #*3 because of RGB
black_img[:, :, 0] = imgs["A"]#red channel
black_img[:, :, 2] = imgs["B"] #blue channel
```

```python
#normalization
max_val = np.max(black_img)
min_val = np.min(black_img)
black_img_norm = (black_img-min_val)/(max_val-min_val)
```

```python
show_img(black_img_norm)
```

```python
#create image
black_img = np.zeros((imgs["A"].shape[0], imgs["A"].shape[1], 3)) #*3 because of RGB
black_img[:, :, 0] = imgs["A"]#red channel
black_img[:, :, 2] = imgs["C"] #blue channel

#normalization
max_val = np.max(black_img)
min_val = np.min(black_img)
black_img_norm = (black_img-min_val)/(max_val-min_val)

#plot image
show_img(black_img_norm)
```

<b><u> Interpretation </u></b> <br>
The normalization (along with other possibilities we tried) did not yield the in the problem sheet described colors. However, based on variation in the color we are able to form ideas about the particle species relationships.
For the particel species A and B the normalized "overlapped" the images main colour is pink as well as dark pruprle (or black). With little red dots. This leads us to the conclusion, that theese particle species might be significantly correlated.
For the second pair (A and C) there is a lot more blue and red, an way less pink and darker regions. This result suggests, that there might be no correlation between A and C or it is very weak. 

**subtask 3**

```python
#flatten the tree initial images
imgA_fl = np.ravel(imgs["A"])
imgB_fl = np.ravel(imgs["B"])
imgC_fl = np.ravel(imgs["C"])
```

```python
fig, ax = plt.subplots()
ax.plot(imgB_fl, imgA_fl, "o", color="darkgreen")
ax.set_xlabel("particle species B")
ax.set_ylabel("particle species A")
ax.set_title("Relationship between particle species A and B")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(imgC_fl, imgA_fl, "o", color="darkgreen")
ax.set_xlabel("particle species C")
ax.set_ylabel("particle species A")
ax.set_title("Relationship between particle species A and C")
plt.show()
```

<b><u> Interpretation </u></b> <br>
The scatterplots confirm the hypothesis we made in subtask 2. For the scatterplot of particle species A and B, the points are on a diagonal with little to no outliers (depending on the conditions of what is an outlier). The second scatter plot, showing the relationship between particle species A and C, does not show a diagonal form. However, it is also not totally square (which as the task states would be the case of beeing uncorrelated), which could mean, that there is a weak correlation. But in general, the second scatterplot confirm our ideas from subtask 2 for theese particle species as well.
